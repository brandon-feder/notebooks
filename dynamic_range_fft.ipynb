```python
import numpy as np
from numpy.fft import fft2, ifft2
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
N=128
#noise at 1e-3
#sources of interest at 1, 1000, 10000
cmap = np.random.normal(0,np.sqrt(1e-4),(N,N))
cmap [N//3,N//3]+=1.*np.exp(1.j*np.random.uniform(0,2*np.pi))
cmap [N//2,N//2]+=np.sqrt(10000)*np.exp(1.j*np.random.uniform(0,2*np.pi))
cmap [N//2,2*N//3]+=np.sqrt(1000)*np.exp(1.j*np.random.uniform(0,2*np.pi))
rmap = np.abs(cmap**2)
plt.imshow(rmap,norm=LogNorm(vmin=1e-3, vmax=11000))
plt.colorbar()
```

```python
fmap = fft2(cmap)
fmapd = fmap * np.exp(1.j*np.random.normal(0,0.05,(N,N)))
cmapC = ifft2(fmap)
cmapD = ifft2(fmapd)
```
