```
from astropy.time import Time, TimeDelta
```

```
tag2tup = lambda tag : tuple([int(x) for x in [tag[0:2],tag[2:4],tag[4:6],tag[7:9],tag[9:11]]])
tup2isot = lambda tup : '20%02d-%02d-%02dT%02d:%02d:00'%tup 
tup2time = lambda tup : Time(tup2isot(tup),format='isot',scale='utc')
tag2time = lambda tag : tup2time(tag2tup(tag))
time2tag = lambda time : '%02d%02d%02d_%02d%02d'%tuple(time.utc.ymdhms)[:5]
def GetTimeList (tagstart,tagend):
    ts = tup2time(tag2tup(tagstart))
    te = tup2time(tag2tup(tagend))
    t = ts
    l = [t]
    while t<=te:
        t += TimeDelta(60,format='sec')
        l.append(t)
    return l
```

```
l=GetTimeList('201103_1200','201103_1205')
```

```
[(time2tag(t),t.mjd) for t in l]
```
